```python
from time import time
from community_detection import *

#############################################
############## code à executer ##############
#############################################

### Question  2 ###

#théoriquement :
print("###Théoriquement : ###\n\n- dico_reseau() à une complexité quadratique étant donné qu'il y a une boucle avec un appelle de fonction ayant une boucle.\n- create_network() à une complexité linéaire executant une seule boucle.\nDe ce fait, create_network() devrait être plus rapide.\n\n###Expériementalement : ###\n")
#expériementalement :

list_of_friends = [
    'Muriel', 'Yasmine', 'Muriel', 'Joël', 'Yasmine', 'Joël', 'Yasmine',
    'Thomas', 'Joël', 'Nassim', 'Joël', 'Andrea', 'Joël', 'Ali', 'Thomas',
    'Daria', 'Thomas', 'Carole', 'Nassim', 'Andrea', 'Nassim', 'Ali', 'Andrea',
    'Ali', 'Thierry', 'Axel', 'Axel', 'Léo', 'Thierry', 'Léo', 'Léo',
    'Valentin', 'Valentin', 'Andrea'
]

tpm = 0
for i in range(10000):
    tic = time()
    dico_reseau(list_of_friends)
    tpm += ((time()-tic)*1000000)
print("Le temps d'execution de dico_reseau() est de :", round(tpm/10000, 2), 'ns.')

tpm = 0
for i in range(10000):
    tic = time()
    create_network(list_of_friends)
    tpm += ((time()-tic)*1000000)
print("Le temps d'execution de create_network() est de :", round(tpm/10000, 2), 'ns.')

print('Ce qui confirme notre théorie, étant donné que la seconde fonction est plus rapide.')

print('---------------------------------------------------------------------------------------------------------------------------')
### Question 11 ###

#théoriquement :
print("###Théoriquement : ###\n\n- find_community_by_decreasing_popularity() est quadratique.\n- find_community_from_person() est quadratique.\nDe ce fait, elles devraient être d'un temps similaire ou presque.\n\n###Expériementalement : ###\n")
#expériementalement :

network = {
    "Alice" : ["Bob", "Dominique"],
    "Bob" : ["Alice", "Charlie", "Dominique"], 
    "Charlie" : ["Bob"], 
    "Dominique" : ["Alice", "Bob"]
}

tpm = 0
for i in range(10000):
    tic = time()
    find_community_by_decreasing_popularity(network)
    tpm += ((time()-tic)*1000000)
print("Le temps d'execution de find_community_by_decreasing_popularity() est de :", round(tpm/10000, 2), 'ns.')

tpm = 0
for i in range(10000):
    tic = time()
    find_community_from_person(network, 'Alice')
    tpm += ((time()-tic)*1000000)
print("Le temps d'execution de find_community_from_person() est de :", round(tpm/10000, 2), 'ns.')

print('Ce qui confirme notre théorie, étant donné que le temps est similaire ou presque. \nCependant, find_community_from_person() semble plus rapide.')
```
